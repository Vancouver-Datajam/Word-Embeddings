```python
# Author Sevy
#importing...
import pandas as pd
import re
import nltk
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
import os
import codecs
```

```python
#cleaning
nltk.download('stopwords')
nltk.download('wordnet')
def clean_data(data):
  data = data.lower()

  # Remove line breaks
  data = re.sub(r'\n', '', data)

  # Remove puncuation (except for periods)
  data = re.sub(r'[^.\w\s]', '', data)

  #remove single characters
  data = re.sub(r'\b\w\b', '', data)

  #remove numbers
  data = re.sub(r'[0-9]', '', data)

  #tokenization
  data = data.split()

  #stop words
  stop_words = nltk.corpus.stopwords.words("english")
  data = [word for word in data if not word in stop_words]

  #lemmatization #is this going to take too long with large datasets??
  lem = WordNetLemmatizer()
  data = [lem.lemmatize(y) for y in data]

  final_string = ' '.join(data)

  return final_string
```

```python
# For testing using files
# Code from https://github.com/alik604/eminem_lyrics_generator/blob/master/eminem_lyrics_generator_LSTM_%26_HMM_Alik604.ipynb

#!git clone https://github.com/SevyV/data.git
#!ls "data/books"

DIR = "data/books"

temp = ""
t = ""
chars = []
for book in os.listdir(DIR):
    print("Reading ", book)
    with codecs.open(DIR + '/' + book, "rb", "utf-8") as infile:
        temp += infile.read()
        chars.append(len(temp))
        print("Characters read so far " + str(len(temp)))

processed_text = temp
processed_text = clean_data(processed_text)
print(len(temp), ' char count to: ', len(processed_text))

lines = pd.DataFrame(processed_text.split("."), columns=["lines"])
print('pre drop_duplicates', lines.shape)
lines = lines.apply(lambda x: x.str.strip())
lines.drop_duplicates(inplace=True)
print('post drop_duplicates', lines.shape)


lines.head(5).values

```
